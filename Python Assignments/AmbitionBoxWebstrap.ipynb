```python
import bs4
from bs4 import BeautifulSoup
import requests
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
```

```python
contents=requests.get("https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav",headers=headers).text
```

```python
soup=BeautifulSoup(contents,"lxml")
```

```python
soup.text
```

# Getting All Company Name

```python
for cname in soup.find_all("h2",class_="companyCardWrapper__companyName"):
    print(cname.text.lstrip())
```

# Getting Rating of all Companies

```python
for crating in soup.find_all("span",class_="companyCardWrapper__companyRatingValue"):
    print(crating.text)
```

# Getting Company Details

```python
for cdetails in soup.find_all("span",class_="companyCardWrapper__interLinking"):
    print(cdetails.text.strip())
```

```python
for cjobs in soup.find_all("span",class_="companyCardWrapper__ActionTitle"):
    print(cjobs.text)
    for cntjob in soup.find_all("span",class_="companyCardWrapper__ActionCount"):
        if cjobs.text=='Jobs':
            print(cjobs.text,cntjob.text)
```
