Q8. WAP to take an int from user. Print number of 1's and 0's in binary format of given number.

NOTE: '0b' is suffix in binary string. It should not be included in calculation.

```python
num=int(input("Enter a number"))
n=bin(num)
print(n)
zero=-1
one=0
for i in n :
  if i=="0" :
    zero=zero+1
  if i=="1" :
    one=one+1
print("The Number of 0 : ",zero)
print("The Number of 1 : ",one)


#using list comprehension
cnt0=[ele for ele in n if ele =="0" ]
cnt1=[ele for ele in n if ele =="1" ]
print("Total zeros:",len(cnt0)-1)
print("Total One:",len(cnt1))
```

Q16. WAP to check given integer is a perfect square or not.
Don't use built in function



```python
num=int(input("Enter a Number"))
for i in range (1,num):
  if num%i==0:
    if num/i==i:
      print(num," is perfect square Number")
      break
else:
  print(num," is not perfect squre Number")
```
